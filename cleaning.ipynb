```python
def get_discount_percent(deal_prices, list_prices):
    all_discount_percent = []
    zipped = zip(deal_prices, list_prices)
    zipped_list = list(zipped)
    try:
        for tuple in zipped_list:
            discount = (tuple[1] - tuple[0])/tuple[1]
            all_discount_percent.append(discount)
    except:
        discount = int(0)
        all_discount_percent.append(discount)
    return all_discount_percent
```

```python
# Convert all of the non- Clothing, Shoes & Jewelry category seller rankings to NaN

def obtain_clothing_category_only(all_product_listings):
    seller_rank_confirmed = []
    for ranking in all_product_listings['seller_rank']:
        if ranking == None:
            seller_rank_confirmed.append(None)
        elif ranking.find('in Clothing, Shoes & Jewelry (See Top 100 in Clothing, Shoes & Jewelry)') < 0:
            seller_rank_confirmed.append(None)
        else:
            seller_rank_confirmed.append(ranking)
    return seller_rank_confirmed
```

```python
def clean_clothing_seller_rank_string(all_product_listings):
    clothing_seller_rank = []
    for seller_rank_string in all_product_listings['seller_rank_confirmed_long']:
        if seller_rank_string == None:
            seller_rank = 0
            clothing_seller_rank.append(seller_rank)
        else:
            ending_string = ' in Clothing, Shoes & Jewelry (See Top 100 in Clothing, Shoes & Jewelry)'
            seller_rank = seller_rank_string[:-len(ending_string)]
            clothing_seller_rank.append(int(seller_rank.replace(',', '')))
    return clothing_seller_rank
```

# Make a log version of the price for possible easier modeling

```python
def get_log_list_prices(all_product_listings_for_modeling):
    log_list_prices = []
    for price in all_product_listings_for_modeling['list_price']:
        log_list_prices.append(math.log(price))
    return log_list_prices
    
```

```python
all_product_listings_for_modeling['log_price'] = get_log_list_prices(all_product_listings_for_modeling)
```
